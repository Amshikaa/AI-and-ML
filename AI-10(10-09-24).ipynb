```python
# Gradient Descent to solve f(x) = x^2 - 25 = 0

# Function f(x)
def f(x):
    return x**2 - 25

# Derivative of f(x)
def df(x):
    return 2 * x

# Gradient Descent algorithm
def gradient_descent(learning_rate, initial_guess, iterations):
    x = initial_guess  # Starting point
    for i in range(iterations):
        gradient = df(x)  # Calculate the gradient at x
        x = x - learning_rate * gradient  # Update the value of x
        print(f"Iteration {i+1}: x = {x}, f(x) = {f(x)}")
    return x

# Parameters
learning_rate = 0.1  # Step size
initial_guess = 10  # Initial guess
iterations = 50  # Number of iterations

# Apply gradient descent
root = gradient_descent(learning_rate, initial_guess, iterations)
print(f"Found root: x = {root}, f(x) = {f(root)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function f(x)
def f(x):
    return x**2 - 25

# Derivative of f(x)
def df(x):
    return 2 * x

# Gradient Descent algorithm
def gradient_descent(learning_rate, initial_guess, iterations):
    x = initial_guess  # Starting point
    x_history = []  # Store the history of x values for plotting
    for i in range(iterations):
        gradient = df(x)  # Calculate the gradient at x
        x = x - learning_rate * gradient  # Update the value of x
        x_history.append(x)  # Store the updated x value for plotting
        print(f"Iteration {i+1}: x = {x}, f(x) = {f(x)}")
    return x, x_history

# Parameters
learning_rate = 0.1  # Step size
initial_guess = 10  # Initial guess
iterations = 50  # Number of iterations

# Apply gradient descent
root, x_history = gradient_descent(learning_rate, initial_guess, iterations)

# Plotting the function and gradient descent updates
x_vals = np.linspace(-10, 10, 400)  # x values for plotting the function
y_vals = f(x_vals)  # y values for the function

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label=r'$f(x) = x^2 - 25$', color='blue')  # Plot the function
plt.scatter(x_history, f(np.array(x_history)), color='red', label='Gradient Descent Updates', zorder=5)  # Plot the updates
plt.plot(x_history, f(np.array(x_history)), color='red', linestyle='dashed', alpha=0.5)  # Line connecting updates
plt.axhline(0, color='black',linewidth=1)  # x-axis line
plt.axvline(0, color='black',linewidth=1)  # y-axis line
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent on f(x) = x^2 - 25')
plt.legend()
plt.grid(True)
plt.show()

print(f"Found root: x = {root}, f(x) = {f(root)}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=np.linspace(-10,10,500) 
Y= X**2-25
```

```python
plt.plot(X,Y)
```

```python
# Building the model
beta0=0
beta1=0
beta2=0

L = 0.0001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = beta0 + beta1*X + beta2*X**2  # The current predicted value of Y
    D_beta0 = (-2/n) * sum(Y - Y_pred) # Derivative wrt m
    D_beta1 = (-2/n) * sum(X * (Y - Y_pred)) # Derivative wrt c
    D_beta2= (-2/n)*sum(X**2 * (Y-Y_pred))
    beta0= beta0 - L *D_beta0 
    beta1= beta1 - L *D_beta1 
    beta2= beta2- L *D_beta2
   
    
print (beta0,beta1,beta2)
# Making predictions
Y_pred =  beta0+beta1*X+beta2*X**2
plt.plot(X,Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```
